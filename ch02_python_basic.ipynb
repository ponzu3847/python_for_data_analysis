#### バイト型とUnicode型

```python
val='español'
print(val)

#UTF-8バイト表現に変換
val_utf8=val.encode('utf-8')
print(val_utf8)
print(type(val_utf8))

#Unicode文字列に戻す
val_utf8=val_utf8.decode('utf-8')
print(val_utf8)
```

```python
#様々なエンコーディング
val='español'
print(val.encode('latin1'))
print(val.encode('utf-16'))
print(val.encode('utf-16le'))
```

```python
#バイト文字列自体の定義(文字列の前に b をつける)
bytes_val=b'this is bytes'
print(bytes_val)
print(type(bytes_val))

decoded=bytes_val.decode('utf-8')
print(decoded)
print(type(decoded))
```

#### 2.3.2.7 日付と時刻

```python
from datetime import datetime,date,time

dt=datetime(2011,10,29,20,30,21)
print(dt.year)
print(dt.month)
print(dt.day)
print(dt.hour)
print(dt.minute)
print(dt.second)
```

```python
dt=datetime(2011,10,29,20,30,21)

print(dt.date())
print(dt.time())
```

```python
dt=datetime(2011,10,29,20,30,21)

dt.strftime('%m/%d/%Y %H:%M')
```

```python
datetime.strptime('20091031','%Y%m%d')
```

```python
dt=datetime(2011,10,29,20,30,21)

dt.replace(minute=0,second=0)
```

```python
#2つのdatatimeオブジェクトの差をとると、結果はdatatime.timedelta型のオブジェクトになる

dt=datetime(2011,10,29,20,30,21)
dt2=datetime(2011,11,15,22,30)

delta=dt2-dt

delta   #出力の(17,7179)は17日と7179秒を表す
```

```python
print(delta)
print(type(delta))
```

```python
#datatimeオブジェクトにtimedeltaオブジェクトを加えると、その分だけ変化した時刻を得る事ができる
dt+delta
```

### 2.3.3 制御フロー

#### 2.3.3.2 forループ

```python
#breakが終了させるのはその最も内側のループ処理だけ(外側の処理は継続する)
for i in range(4):
    for j in range(4):
        if j>i:
            break
        print((i,j))
```
